# AI-Based Market Validation Trends Dashboard

```python
import requests
import pandas as pd
import plotly.express as px
```

```python
# dataset of gitHub repo
url = "https://api.github.com/search/repositories?q=AI"
```

```python
response = requests.get(url)
print(response)
```

```python
data = response.json()
repos = data['items']
print(repos)
```

```python
df = pd.DataFrame(repos)
df.head()
```

```python
df.info()
```

```python
df = df[['name', 'stargazers_count', 'forks_count', 'created_at']]
df.tail()
```

```python
df['created_at'] = pd.to_datetime(df['created_at'])
df['created_at'].tail()
```

```python
# bar chart..
fig = px.bar(df, x='created_at', y='stargazers_count', title='GitHub Stars Over Time')
fig.show()
```

```python
# Line chart trend over time..
fig = px.line(df, x='created_at', y='stargazers_count', title='GitHub Stars Over Time')
fig.show()
```
